# Ajustando a base

```python
#Renomeando as colunas
nova_base = base_mkt.rename(columns={'youtube': 'Sales', 'facebook': 'YouTube', 'newspaper': 'Facebook', 'sales': 'Newspaper'})
nova_base.head(10)
```

# Agora que a base est√° correta, vamos adicionar uma coluna de Investimento e ROI

```python
#Criando nova coluna 'investimento' e 'ROI'
nova_base_att = nova_base
nova_base_att['Investimento'] = nova_base_att['YouTube'] + nova_base_att['Facebook'] + nova_base_att['Newspaper']
nova_base_att['ROI'] = round((nova_base_att['Sales'] - nova_base_att['Investimento']) / nova_base_att['Investimento'], 2)
nova_base_att.head()
```

```python
#Gr√°fico de linhas por tipo de investimento
multi = nova_base_att[['YouTube', 'Facebook', 'Newspaper']]
plt.figure(figsize=(12,8))
plt.style.use('seaborn-v0_8-dark')
palette = plt.get_cmap('Paired')
num = 0
for col in multi:
  num += 1
  plt.subplot(3, 1, num)
  for i in multi:
    plt.plot(multi.index, multi[i], marker='', color='red', linewidth=0.8, alpha=0.2)
  plt.plot(multi.index, multi[col], marker='', color=palette(num), linewidth=1.6, alpha=0.9, label=col)
  plt.title(col, loc='center', fontsize=10, fontweight=2, color='black')
plt.show();
```

```python
#Gr√°fico de linhas de total de investimento x vendas
fig, ay = plt.subplots(figsize=(16, 8))
sns.lineplot(data=nova_base_att, x=nova_base_att.index, y='Investimento', color='yellow', label='Investimento')
sns.lineplot(data=nova_base_att, x=nova_base_att.index, y='Sales', color='green', label='Sales')
ay.fill_between(
    nova_base_att.index, nova_base_att['Investimento'], nova_base_att['Sales'],
    where=(nova_base_att['Investimento'] > nova_base_att['Sales']),
    interpolate=True, color='yellow', alpha=0.25
    )
ay.fill_between(
    nova_base_att.index, nova_base_att['Investimento'], nova_base_att['Sales'],
    where=(nova_base_att['Investimento'] < nova_base_att['Sales']),
    interpolate=True, color='green', alpha=0.25
    )
ay.legend()
plt.xlabel('Qtd')
plt.ylabel('Valor')
plt.title('Valor Total de Investimento e Vendas');
plt.show()
```

# Regress√£o Linear
Agroa vamos treinar um modelo e aplicar uma regress√£o linear simples.

```python
nova_base_att.columns
```

```python
#Separando Variaveis X e Y.
X = nova_base_att[['YouTube', 'Facebook', 'Newspaper']]

Y = nova_base_att['Sales']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.8, random_state=22)
```

```python
# Chamando a regress√£o e treinando o modelo.
lm = LinearRegression()
lm.fit(X_train, Y_train)
```

```python
#TEstando Modelo
modelo_teste = lm.predict(X_test)
```

```python
#Analisando.
precisao = r2_score(Y_test, modelo_teste)
print(f'A precis√£o deste modelo √© de {round(precisao*100)}%.')
```
